```python
#question 1
def pentagonal(n):
    
    return (5 * n * n - 5 * n + 2) / 2
print(pentagonal(2))
```

```python
#question 2
def encrypt(s):
    d = {'a':0,'e':1,'i':2,'o':2,'u':3}
    c = ''
    s1 = ''
    for i in s:
        s1 = i+s1
    for i in s1:
        if i in d:
            c = c+str(d[i])
        else:
            c = c+i
    return c+'aca'
encrypt('apple')
```

```python
# question 3

import datetime
from datetime import datetime
import calendar

def has_friday_13(month, year):
    date = f'{month} 13 {year}'.format()
    day = datetime.strptime(date, '%m %d %Y').weekday()
    return (calendar.day_name[day]) == "Friday"
print(has_friday_13(1,1985))
```

```python
# question 5
def pluralize(a):
    l = []
    for i in a:
        if a.count(i)>1:
            i = i+'s'
        l.append(i)
    return set(l)
print(pluralize(["cow", "pig", "cow", "cow"]))
```
